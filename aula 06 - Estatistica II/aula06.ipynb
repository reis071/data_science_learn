## Oque √â Probabilidade?

Probabilidade √© a √°rea da matem√°tica que estuda a chance de um evento acontecer. Imagine que voc√™ joga uma moeda para o alto. Como a moeda tem dois lados, cara ou coroa, a probabilidade de cair um desses lados √© de 50%.

### Como Calcular a Probabilidade

$$
P(E) = \frac{\text{N√∫mero de resultados favor√°veis}}{\text{N√∫mero total de resultados poss√≠veis}}
$$

ùëÉ(ùê∏) √© a probabilidade do evento ùê∏ acontecer.

N√∫mero de resultados favor√°veis √© o n√∫mero de maneiras que o evento pode acontecer.

N√∫mero total de resultados poss√≠veis √© o n√∫mero total de todos os eventos poss√≠veis.

### Exemplo

Suponha que voc√™ tem um dado comum com 6 faces numeradas de 1 a 6. Qual √© a probabilidade de voc√™ tirar um n√∫mero 4?

N√∫mero de resultados favor√°veis: 1 (apenas o n√∫mero 4)
N√∫mero total de resultados poss√≠veis: 6 (j√° que o dado tem 6 faces)

$$
P(4) = \frac{1}{6} \approx 0.167
$$

Portanto, a probabilidade de tirar o n√∫mero 4 √© cerca de 16,7%.

```python
import pandas as pd

# Suponha que voc√™ tem o seguinte DataFrame com os resultados dos lan√ßamentos de um dado
data = {'Lancamento': [1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 6, 6]}
df = pd.DataFrame(data)

# Calcular a probabilidade de tirar um n√∫mero 6
total_lancamentos = len(df)
resultados_favoraveis = df[df['Lancamento'] == 6].count()

probabilidade_6 = resultados_favoraveis / total_lancamentos

print(f"A probabilidade de tirar um n√∫mero 6 √©: {probabilidade_6[0]:.2f}")
```

## Graficos

Matplotlib √© uma biblioteca em Python usada para criar visualiza√ß√µes gr√°ficas. √â uma das bibliotecas mais populares para gerar gr√°ficos, e √© amplamente utilizada em ci√™ncia de dados, aprendizado de m√°quina e an√°lise de dados. Gr√°ficos s√£o representa√ß√µes visuais de dados que ajudam a entender, analisar e comunicar informa√ß√µes de forma clara e eficiente. Ao inv√©s de olhar para uma tabela cheia de n√∫meros, um gr√°fico pode tornar a informa√ß√£o mais acess√≠vel, mostrando padr√µes, tend√™ncias ou compara√ß√µes de forma imediata.

### Tipos de Gr√°ficos

#### Histograma

Um histograma √© um tipo de gr√°fico que mostra a distribui√ß√£o de uma vari√°vel cont√≠nua. Ele agrupa os dados em intervalos (bins) e representa a frequ√™ncia ou a contagem de valores que caem em cada intervalo. Diferente do gr√°fico de barras, que compara categorias, o histograma √© usado para visualizar a distribui√ß√£o de dados e identificar padr√µes, como a forma (normal, assim√©trica, etc.), a dispers√£o, e a presen√ßa de outliers.

#### Como Funciona?

**Eixos**: No histograma, o eixo horizontal (x) representa os intervalos de valores, e o eixo vertical (y) mostra a frequ√™ncia ou a contagem de valores dentro de cada intervalo.

**Barras**: Cada barra no histograma representa o n√∫mero de ocorr√™ncias (frequ√™ncia) de dados dentro de um intervalo espec√≠fico

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados de exemplo: idades de um grupo de pessoas
idades = [22, 25, 28, 22, 30, 32, 35, 40, 28, 30, 32, 35, 40, 45, 50, 52, 55, 60, 22, 28, 35, 40, 50, 60]

# Criar um DataFrame
df = pd.DataFrame(idades, columns=['Idade'])

# Criar o histograma
df['Idade'].plot(kind='hist', bins=5, color='skyblue', edgecolor='black')

# Adicionar t√≠tulo e r√≥tulos
plt.title('Distribui√ß√£o de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequ√™ncia')
plt.grid(axis='y')
plt.show()
```

### Gr√°fico De Barras

Um gr√°fico de barras √© uma forma visual de comparar valores entre diferentes categorias ou grupos. Nele, cada barra representa uma categoria, e a altura (ou comprimento) da barra indica o valor associado a essa categoria. √â especialmente √∫til quando voc√™ deseja comparar diferentes itens ou ver mudan√ßas em uma √∫nica vari√°vel ao longo do tempo.

#### Como Funciona?

**Eixos**: O eixo horizontal (x) geralmente representa as categorias, e o eixo vertical (y) mostra os valores (como contagem, m√©dia, porcentagem, etc.).

**Barras**: Cada barra corresponde a uma categoria. Quanto maior a barra, maior o valor que ela representa.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados de exemplo: Vendas de diferentes produtos
dados = {
    'Produto': ['Produto A', 'Produto B', 'Produto C', 'Produto D'],
    'Vendas': [250, 150, 300, 200]
}

# Criar um DataFrame
df = pd.DataFrame(dados)

# Criar o gr√°fico de barras
df.plot(kind='bar', x='Produto', y='Vendas', color='lightgreen', edgecolor='black')

# Adicionar t√≠tulo e r√≥tulos aos eixos
plt.title('Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.grid(True, linestyle='--', alpha=0.6)

# Exibir o gr√°fico
plt.show()
```

### Gr√°fico De Pizza

O gr√°fico de pizza (ou gr√°fico de setores) √© uma representa√ß√£o visual usada para mostrar a propor√ß√£o de partes em rela√ß√£o ao todo. Cada "fatia" da pizza representa uma categoria, e o tamanho da fatia indica a porcentagem ou propor√ß√£o daquela categoria em rela√ß√£o ao total. Este tipo de gr√°fico √© √∫til para mostrar a distribui√ß√£o de um conjunto de dados de maneira intuitiva.

#### Como Funciona? 

**Fatia**: Cada fatia representa uma categoria e seu tamanho corresponde √† propor√ß√£o daquela categoria no conjunto de dados.

**Total**: O gr√°fico de pizza sempre representa 100% do total, e as fatias somadas devem dar esse valor.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados de exemplo: Participa√ß√£o de mercado de diferentes empresas
dados = {
    'Empresa': ['Empresa A', 'Empresa B', 'Empresa C', 'Empresa D'],
    'Participa√ß√£o (%)': [35, 25, 20, 20]
}

# Criar um DataFrame
df = pd.DataFrame(dados)

# Criar o gr√°fico de pizza
plt.pie(df['Participa√ß√£o (%)'], labels=df['Empresa'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])

# Adicionar t√≠tulo
plt.title('Participa√ß√£o de Mercado por Empresa')

# Exibir o gr√°fico
plt.show()
```

### Intervalo Interquart√≠lico (IQR)

O Intervalo Interquart√≠lico (IQR) √© uma medida de dispers√£o que descreve a amplitude do intervalo central de um conjunto de dados, eliminando a influ√™ncia de valores extremos ou outliers. Ele √© definido como a diferen√ßa entre o terceiro quartil (Q3) e o primeiro quartil (Q1) de um conjunto de dados.

#### Quartis 

Q1 (Primeiro Quartil): Representa o valor abaixo do qual se encontram 25% dos dados. √â tamb√©m conhecido como o 25¬∫ percentil.

Q2 (Mediana): O segundo quartil, ou mediana, divide o conjunto de dados ao meio, com 50% dos dados abaixo dele.

Q3 (Terceiro Quartil): Representa o valor abaixo do qual se encontram 75% dos dados. √â tamb√©m conhecido como o 75¬∫ percentil.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados de exemplo: notas de um grupo de estudantes
notas = [55, 70, 75, 80, 85, 90, 95, 100, 105, 110]

# Criar um DataFrame
df = pd.DataFrame(notas, columns=['Notas'])

# Criar o boxplot
plt.figure(figsize=(8, 6))
boxplot = df.boxplot(column=['Notas'], grid=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Calcular Q1, Q2 (mediana) e Q3
Q1 = df['Notas'].quantile(0.25)
Q2 = df['Notas'].median()
Q3 = df['Notas'].quantile(0.75)

# Adicionar linhas horizontais para os quartis
plt.axhline(y=Q1, color='orange', linestyle='--', linewidth=2, label='Q1 (1¬∫ Quartil)')
plt.axhline(y=Q2, color='green', linestyle='--', linewidth=2, label='Q2 (Mediana)')
plt.axhline(y=Q3, color='red', linestyle='--', linewidth=2, label='Q3 (3¬∫ Quartil)')

# Adicionar t√≠tulo e r√≥tulos aos eixos
plt.title('Boxplot das Notas dos Estudantes com Quartis Sinalizados')
plt.ylabel('Notas')

# Adicionar legenda para identificar os quartis
plt.legend()

# Exibir o gr√°fico
plt.show()
```

## Distribui√ß√£o Normal

A distribui√ß√£o normal, tamb√©m conhecida como distribui√ß√£o Gaussiana ou curva de sino, √© uma das distribui√ß√µes de probabilidade mais importantes em estat√≠stica. Ela descreve como os valores de uma vari√°vel se distribuem de maneira sim√©trica em torno da m√©dia. Muitos fen√¥menos naturais, como altura, peso, e notas de testes, tendem a seguir uma distribui√ß√£o normal.

### Caracter√≠sticas da Distribui√ß√£o Normal

**Simetria**: A curva √© sim√©trica em torno da m√©dia, o que significa que a m√©dia, a mediana e a moda s√£o iguais.
M√©dia (Œº): O valor central da distribui√ß√£o.

**Desvio Padr√£o (œÉ)**: Indica a dispers√£o dos dados em rela√ß√£o √† m√©dia. Na distribui√ß√£o normal, aproximadamente 68% dos dados caem dentro de 1 desvio padr√£o da m√©dia, 95% dentro de 2 desvios padr√£o, e 99.7% dentro de 3 desvios padr√£o.

**Curva em Sino**: A forma da curva da distribui√ß√£o normal se assemelha a um sino, com a maioria dos valores concentrados em torno da m√©dia.

### Min e Max

Os valores m√≠nimo (min) e m√°ximo (max) s√£o os extremos de um conjunto de dados. Eles representam, respectivamente, o menor e o maior valor presente no conjunto.

Min (m√≠nimo): √â o menor valor de um conjunto de dados.

Max (m√°ximo): √â o maior valor de um conjunto de dados.

Estes valores s√£o frequentemente utilizados para descrever o intervalo ou amplitude dos dados, que √© simplesmente a diferen√ßa entre o m√°ximo e o m√≠nimo

### Z-Score

O Z-Score (ou escore z) √© uma medida estat√≠stica que descreve a posi√ß√£o de um valor em rela√ß√£o √† m√©dia de um conjunto de dados, em termos de desvios padr√£o. Ele indica quantos desvios padr√£o um determinado valor est√° acima ou abaixo da m√©dia.

### Visualizando uma Distribuicao Normalizada

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gerar dados simulados de uma distribui√ß√£o normal
mu, sigma = 100, 15  # m√©dia e desvio padr√£o
dados = np.random.normal(mu, sigma, 1000)

# Criar um DataFrame
df = pd.DataFrame(dados, columns=['Valores'])

# Plotar o histograma
plt.figure(figsize=(10, 6))
plt.hist(df['Valores'], bins=30, color='lightblue', edgecolor='black', density=True)

# Plotar a curva da distribui√ß√£o normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-((x - mu) ** 2) / (2 * sigma ** 2)) / (sigma * np.sqrt(2 * np.pi))
plt.plot(x, p, 'r', linewidth=2)

# Adicionar t√≠tulo e r√≥tulos aos eixos
plt.title('Histograma da Distribui√ß√£o Normal Simulada')
plt.xlabel('Valores')
plt.ylabel('Densidade')

# Exibir o gr√°fico
plt.show()
```

#### Explica√ß√£o

Gerar Dados: Usamos np.random.normal(mu, sigma, 1000) para gerar 1000 valores que seguem uma distribui√ß√£o normal com m√©dia ùúá =100  e desvio padr√£o ùúé = 15.

Histograma: O histograma mostra a distribui√ß√£o dos dados gerados. O par√¢metro density=True normaliza o histograma para que a √°rea total seja igual a 1.

Curva da Distribui√ß√£o: A linha vermelha sobreposta no histograma representa a curva te√≥rica da distribui√ß√£o normal para os mesmos ùúá e ùúé, o que ajuda a visualizar a conformidade dos dados simulados com a distribui√ß√£o normal.
Customiza√ß√£o: Adicionamos t√≠tulo e r√≥tulos para melhorar a legibilidade do gr√°fico.

### F√≥rmula da Distribui√ß√£o Normal

A fun√ß√£o de densidade de probabilidade para a distribui√ß√£o normal √© dada por:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

Onde:
- \( x \) √© o valor da vari√°vel aleat√≥ria.
- \( \mu \) √© a m√©dia da distribui√ß√£o.
- \( \sigma \) √© o desvio padr√£o.
- \( \exp \) representa a fun√ß√£o exponencial.

Essa f√≥rmula mostra como a probabilidade se distribui em torno da m√©dia \( \mu \) para uma vari√°vel que segue uma distribui√ß√£o normal, sendo mais alta pr√≥ximo √† m√©dia e caindo conforme nos afastamos dela.
